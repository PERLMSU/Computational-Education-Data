# Predictors of Computation in introductory Courses

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
from scipy import stats
```

```python
#importing the uncleaned data
data_uncleaned=pd.read_csv('Computational Data Without Identifiers.csv')
#importing the cleaned data
data=pd.read_csv(r'C:\Users\Grant\Desktop\Cleaned PERL Research data.csv', encoding='latin1')
```

## Required vs Elective

```python
#This is for column J
Intro_Course_Required = []
for row in data.values[:,9]:
    Intro_Course_Required.append(row)
Intro_Course_Required_Y = Intro_Course_Required.count("Checked")
Intro_Course_Required_N = Intro_Course_Required.count("Unchecked")
Prop_Intro_Course_Required_Y = Intro_Course_Required_Y/(Intro_Course_Required_Y+Intro_Course_Required_N)
Prop_Intro_Course_Required_N = Intro_Course_Required_N/(Intro_Course_Required_Y+Intro_Course_Required_N)
#This is for column K
Intro_Course_Elective = []
for row in data.values[:,10]:
    Intro_Course_Elective.append(row)
Intro_Course_Elective_Y = Intro_Course_Elective.count("Checked")
Intro_Course_Elective_N = Intro_Course_Elective.count("Unchecked")
Prop_Intro_Course_Elective_Y = Intro_Course_Elective_Y/(Intro_Course_Elective_Y+Intro_Course_Elective_N)
Prop_Intro_Course_Elective_N = Intro_Course_Elective_N/(Intro_Course_Elective_Y+Intro_Course_Elective_N)
```

```python
Comp_Dir = ['In a Required Course', 'Not In a Required Course', 'In an Elective', 'Not In an Elective']
Tick = [1,2,3,4]
Comp_Dir_dist = [Prop_Intro_Course_Required_Y, Prop_Intro_Course_Required_N, Prop_Intro_Course_Elective_Y, Prop_Intro_Course_Elective_N]
plt.xticks(Tick, Comp_Dir, rotation=45)
plt.bar(Tick,Comp_Dir_dist)
plt.title('Of Those who Teach Computation in Introductory Courses, Where is it Taught?')
plt.ylabel('Proportion')
```

## What Programs Does Your University Offer in Computational Physics
### A BS/BA

```python
#Lists for Offering a BS/BA in Computation and if computation is taught in intro courses
Deg_Comp_Phy_YN = []
#Collumn C
for row in data.values[:,2]:
    Deg_Comp_Phy_YN.append(row)
Taught_in_Intro_YN = []
for row in data.values[:,8]:
    Taught_in_Intro_YN.append(row)
Deg_Comp_Y_and_Taught_Intro_Y=0
Deg_Comp_Y_and_Taught_Intro_N=0
Deg_Comp_N_and_Taught_Intro_Y=0
Deg_Comp_N_and_Taught_Intro_N=0
for i,j in zip(Deg_Comp_Phy_YN, Taught_in_Intro_YN):
    if i == "Checked" and j == "Checked":
        Deg_Comp_Y_and_Taught_Intro_Y+=1
    if i == "Checked" and j == "Unchecked":
        Deg_Comp_Y_and_Taught_Intro_N+=1
    if i == "Unchecked" and j == "Checked":
        Deg_Comp_N_and_Taught_Intro_Y+=1
    if i == "Unchecked" and j == "Unchecked":
        Deg_Comp_N_and_Taught_Intro_N+=1
Prop_Deg_Comp_Y_and_Taught_Intro_Y = Deg_Comp_Y_and_Taught_Intro_Y/(Deg_Comp_Y_and_Taught_Intro_Y+Deg_Comp_Y_and_Taught_Intro_N)
Prop_Deg_Comp_Y_and_Taught_Intro_N = Deg_Comp_Y_and_Taught_Intro_N/(Deg_Comp_Y_and_Taught_Intro_Y+Deg_Comp_Y_and_Taught_Intro_N)
Prop_Deg_Comp_N_and_Taught_Intro_Y = Deg_Comp_N_and_Taught_Intro_Y/(Deg_Comp_N_and_Taught_Intro_Y+Deg_Comp_N_and_Taught_Intro_N)
Prop_Deg_Comp_N_and_Taught_Intro_N = Deg_Comp_N_and_Taught_Intro_N/(Deg_Comp_N_and_Taught_Intro_Y+Deg_Comp_N_and_Taught_Intro_N)
```
