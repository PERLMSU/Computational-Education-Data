## Question 8 and 9

```python
Soft_Access_YN = []
#Collumn AS
for row in data.values[:,44]:
    Soft_Access_YN.append(row)
Soft_Access_Y = Soft_Access_YN.count("Yes") 
Prop_Soft_Access_Y = Soft_Access_Y/(Soft_Access_Y+Soft_Access_YN.count("No"))
Prop_Soft_Access_N = 1 - Prop_Soft_Access_Y
HPC_Access_YN = []
#Collumn AV
for row in data.values[:,47]:
    HPC_Access_YN.append(row)
HPC_Access_Y = HPC_Access_YN.count("Yes") 
Prop_HPC_Access_Y = HPC_Access_Y/(HPC_Access_Y+HPC_Access_YN.count("No"))
Prop_HPC_Access_N = 1 - Prop_HPC_Access_Y
```

```python
Compare_8_to_9 = ['Yes', 'No', 'Yes', 'No']
Tick_Compare_8_to_9 = [1,2,3,4]
Compare_8_to_9_dist = [Prop_Soft_Access_Y, Prop_Soft_Access_N, Prop_HPC_Access_Y, Prop_HPC_Access_N]
plt.xticks(Tick_Compare_8_to_9, Compare_8_to_9, rotation=45)
plt.bar(Tick_Compare_8_to_9, Compare_8_to_9_dist)
plt.title('Comparison of Student Access to university Facilities')
plt.ylabel('Proportion')
plt.xlabel('Worstations With Installed Software vs. HPCs')
```
